```python
import matplotlib.pyplot as plt
sns.countplot(x='home_ownership', data=data)
plt.show()
sns.countplot(x='income_category', data=data)
plt.show()
sns.countplot(x='grade', data=data)
plt.show()
sns.countplot(x='region', data=data)
plt.show()
```

```python
sns.countplot(data=data,x='home_ownership')
plt.xticks(rotation=45)
plt.title("Home ownership")
plt.show()
```

```python
plt.pie(data['income_category'].value_counts(),labels=data['income_category'].value_counts().index,autopct='%1.1f%%')
plt.title('Income Category')
plt.show()
```

```python
plt.pie(data['region'].value_counts(),labels=data['region'].value_counts().index,autopct='%1.1f%%')
plt.title('Region')
plt.show()
```

```python
df=data.copy()
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lc=LabelEncoder()
df['region']=lc.fit_transform(df['region'])
```

```python
df.drop(['id', 'home_ownership', 'income_category','purpose','grade','loan_condition','term','interest_payments','application_type','application_type_cat','issue_d','final_d'],axis=1, inplace=True)
```

```python
df.dtypes
```

```python
df
```

```python
corr=df.corr()
plt.figure(figsize=(20,20))
sns.heatmap(corr,annot=True,cmap='coolwarm',linewidths=2)
plt.title('Correlation Heat Map')
plt.show()
```

```python
import matplotlib.pyplot as plt
sns.boxplot(data=df, x="grade_cat", y="interest_rate", hue="grade_cat")
plt.title("Interest Rate vs Grade Category")
plt.show()
```

```python
sns.scatterplot(data=df, x="loan_amount", y="installment")
plt.title("loan_amount vs installment")
plt.show()
```

```python
sns.scatterplot(data=df, x="total_pymnt", y="installment")
plt.title("total_pymnt  vs installment")
plt.show()
```
